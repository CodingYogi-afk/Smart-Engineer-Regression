```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
data = pd.read_csv('train_data.csv')
target = pd.DataFrame(data, columns=['E'])
features = data.drop(['E'],axis=1)
data.describe()
```

```python
T_IDX=0
V_IDX=1
P_IDX=2
RH_IDX=3

machines_stats = features.mean().values.reshape(1, 4)
```

```python
features.mean().values.reshape(1, 4)
```

```python
regr = LinearRegression().fit(features, target)
fitted_vals = regr.predict(features)

MSE = mean_squared_error(target, fitted_vals)
RMSE = np.sqrt(MSE)
```

```python
def get_estimate(Tem, Vac, Pre, Hum):
    
    machines_stats[0][T_IDX] = Tem
    machines_stats[0][V_IDX] = Vac
    machines_stats[0][P_IDX] = Pre
    machines_stats[0][RH_IDX] = Hum
    
    energy_estimate = regr.predict(machines_stats)[0][0]
    
    return energy_estimate    
    
```

```python
print(get_estimate(236.34,406.26,91.32,52.36))
print(get_estimate(237.51,308.58,91.85,53.03))
```
